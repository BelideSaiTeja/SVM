```python
class Support_Vector_Machine:
  def __init__(self, visualisation = True):
    self.visualisation = visualisation
    self.colors = {1 : 'r', -1 : 'b'}
    if self.visualisation:
      self.fig = plt.figure()
      self.ax = self.fig.add_subplot(1, 1, 1)
  # train
  def fit(self, data):
    self.data = data
    #  dict blue print { ||w|| : [w, b]}
    opt_dict = {}
    transforms = [[1, 1],
                  [1, -1],
                  [-1, 1],
                  [-1, -1]]    # transforms are multiplied with vectr w as |w| cannot ffected by -ves so,
                               # [5,5], [5,-5], [-5,5], [-5,-5] ==> root(50) but w.x can affect
                               # so to have all types transforms are used
    all_data = []
    for yi in self.data:
      for featureset in self.data[yi]:
        for feature in featureset:
            all_data.append(feature)
    self.max_feature_value = max(all_data)
    self.min_feature_value = min(all_data)
    all_data = None

    # Support Vector yi*(xi.w + b) = 1         yi = +1 for +ve class yi = -1 for -ve class
    # Decision Vector yi*(xi.w + b) = 0

    step_sizes = [self.max_feature_value * 0.1,
                  self.max_feature_value * 0.01,
                  self.max_feature_value * 0.001,
                  ]

    b_range_multiple = 5
    b_multiple = 5
    latest_optimum = self.max_feature_value * 10

    # actual stepping (down he curve to predict value of min(||w||) and max(b) )
    for step in step_sizes:
      w = np.array([latest_optimum, latest_optimum])
      optimized = False       # can be done bcz. of convex nature of 'w'
      while not optimized:
        for b in np.arange(-1*(self.max_feature_value * b_range_multiple),
                           self.max_feature_value * b_range_multiple,
                           step * b_multiple):
          for transformation in transforms:
            w_t = w * transformation
            found_option = True
            # yi * (xi.w + b) >= 1  --> is the actual case
            for i in self.data:
              for xi in self.data[i]:
                yi = i
                if not yi*(np.dot(w_t, xi)+b) >= 1:
                  found_option = False
                #print(xi, ':', yi*(np.dot(w_t, xi)+b) )
            if found_option:
              opt_dict[np.linalg.norm(w_t)] = [w_t, b]
        if w[0] < 0:
          optimized = True
          print('Optimized a step...')
        else:
          w = w - step

      norms = sorted([n for n in opt_dict])
      opt_choice = opt_dict[norms[0]]
      # opt_choice = [w, b]
      self.w = opt_choice[0]
      self.b = opt_choice[1]
      latest_optimum = opt_choice[0][0] + step*2

    '''
    # printing optimization steps
    for i in self.data:
      for xi in self.data[i]:
        yi = i
        print(xi, ':', yi*(np.dot(self.w, xi)+ self.b) )
    '''

  def predict(self, features):
    # based on sign (x.w + b) classification is done
    classification = np.sign(np.dot(np.array(features), self.w) + self.b)    # vector w and scaler b are not known
    if classification != 0 and self.visualisation:
      self.ax.scatter(features[0], features[1], s = 200, marker = '*', c = self.colors[classification])
    return classification
  def visualise(self):
    [[self.ax.scatter(x[0], x[1], s = 100, color = self.colors[i]) for x in data_dict[i]] for i in data_dict]

    # hyperplane = w.x + b
    # v =w.x + b
    # for positive support vector v = 1
    # for negative support vector v = -1
    # for decision vector v = 0
    def hyperplane(x, w, b, v):
      return (-w[0] * x - b + v)/w[1]
    datarange = (self.min_feature_value * 0.9, self.max_feature_value * 1.1)
    hyp_x_min = datarange[0]
    hyp_x_max = datarange[1]

    # positive support vector hyperplane
    # w.x + b = 1
    psv1 = hyperplane(hyp_x_min, self.w, self.b, 1)     # y of hyperplane
    psv2 = hyperplane(hyp_x_max, self.w, self.b, 1)
    self.ax.plot([hyp_x_min, hyp_x_max], [psv1, psv2], 'k')

    # negative support vector hyperplane
    # w.x + b = -1
    nsv1 = hyperplane(hyp_x_min, self.w, self.b, -1)     # y of hyperplane
    nsv2 = hyperplane(hyp_x_max, self.w, self.b, -1)
    self.ax.plot([hyp_x_min, hyp_x_max], [nsv1, nsv2], 'k')

    # decision support vector hyperplane
    # w.x + b = 0
    db1 = hyperplane(hyp_x_min, self.w, self.b, 0)     # y of hyperplane
    db2 = hyperplane(hyp_x_max, self.w, self.b, 0)
    self.ax.plot([hyp_x_min, hyp_x_max], [db1, db2], 'y--')

    plt.show()

data_dict = {-1 : np.array([[1,7], [2,8],[3,8]]),
             1 : np.array([[5,1], [6,-1], [7,3]])}
```

**Introduction of Kernels**

The most interesting feature of SVM is that it can even work with a non-linear dataset and for this, we use “Kernel Trick” which makes it easier to classifies the points.

Kernel majorly converts the lower dimension space to a higher dimension space using some quadratic functions which will allow us to find a decision boundary that clearly divides the data points.

These functions which help us do this are called **Kernels** and which kernel to use is purely determined by **hyperparameter tuning**.

**Some of the Kernels includes**

Polynomial kernel

RBF kernel

Sigmoid kernel

Bessel function kernel

Anova Kernel

**Support Vectors** are data points that are closer to the hyperplane and influence the position and orientation of the hyperplane.

To know whether model is overfiyment, find the #sv --> no. of support vectors and # samples --> total numberof samples

If #sv/#samples >10% then there might be an issue of overfitment and accuracy on test data will be decreased.

In this case.change the kernel type and try for different.



Generally, SVM is binary classifier. When a data has more than two classes with a single decision boundary, it might be helpful.

Here, we have two methods to classify binary data. This includes,



*   OVR (One verse Rest) one group versus rest of the groups
*   OVO (One verse one)


**OVR** is as follows

Here after every group, thereis an imbalance in data i.e., 4 vs 8 data points for every decision boundary.

It is the default which will be used.

```python
linear_svc = svm.SVC(kernel = 'linear')
linear_svc.fit(X_train, y_train)
```

```python
linear_svc.score(X_test, y_test)
```

```python
poly_svc = svm.SVC(kernel = 'poly')
poly_svc.fit(X_train, y_train)
```

```python
poly_svc.score(X_test, y_test)
```

```python
sigmoid_svc = svm.SVC(kernel = 'sigmoid')
sigmoid_svc.fit(X_train,y_train)
```

```python
sigmoid_svc.score(X_test, y_test)
```

```python
print('Accuracy of rbf kernel = ', rbf_svc.score(X_test, y_test))
print('Accuracy of linear kernel = ', linear_svc.score(X_test, y_test))
print('Accuracy of sigmoid kernel = ', sigmoid_svc.score(X_test, y_test))
print('Accuracy of poly kernel = ', poly_svc.score(X_test, y_test))
```

**Advantages of SVM**
1. SVM works better when the data is Linear

2. It is more effective in high dimensions

3. With the help of the kernel trick, we can solve any complex problem

4. SVM is not sensitive to outliers

5. Can help us with Image classification

**Disadvantages of SVM**
1. Choosing a good kernel is not easy

2. It doesn’t show good results on a big dataset

3. The SVM hyperparameters are Cost -C and gamma. It is not that easy to fine-tune these hyper-parameters. It is hard to visualize their impact
